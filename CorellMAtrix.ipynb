```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot

con = pd.read_csv('Data2.csv')
con
```

```python
cormat = con.corr()
cormat = round(cormat,2);
print(cormat)
cormat.to_csv("cormat.csv", index=True, header=True)
```

```python
pyplot.figure(figsize=(16, 16))
sns.heatmap(cormat,xticklabels=1, yticklabels=1)
```
